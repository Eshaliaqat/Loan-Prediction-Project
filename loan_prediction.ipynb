```python
# Gender, Married, Dependents
for col in ['Gender', 'Married', 'Dependents']:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
# LoanAmount
data['LoanAmount'].fillna(data['LoanAmount'].median(), inplace=True)
```

```python
data['Credit_History'].fillna(data['Credit_History'].mode()[0], inplace=True)
```

```python
# Self_Employed missing values fill
data['Self_Employed'].fillna(data['Self_Employed'].mode()[0], inplace=True)
```

```python
# Loan_Amount_Term missing values fill
data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# Gender
data['Gender'] = data['Gender'].map({'Male':1, 'Female':0})

# Married
data['Married'] = data['Married'].map({'Yes':1, 'No':0})

# Education
data['Education'] = data['Education'].map({'Graduate':1, 'Not Graduate':0})

# Self_Employed
data['Self_Employed'] = data['Self_Employed'].map({'Yes':1, 'No':0})

# Dependents → convert '3+' to 3
data['Dependents'] = data['Dependents'].replace('3+', 3).astype(int)

# Property_Area → One Hot Encoding
data = pd.get_dummies(data, columns=['Property_Area'], drop_first=True)

# Target variable Loan_Status
data['Loan_Status'] = data['Loan_Status'].map({'Y':1, 'N':0})
```
